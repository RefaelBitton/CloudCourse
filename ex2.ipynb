```python
from typing import Text
import json
import ipywidgets as widgets
from IPython.display import display
from google.colab import drive

file_path = '/content/drive/MyDrive/students.json'

student_dict = {
    "Refael Bitton": {
        "first_name": "Refael",
        "last_name": "Bitton",
        "email": "refael.bitton@e.braude.ac.il",
        "courses": "WEB, פיזיקה, מבוא לפיזיקה, מחשוב ענן, קומפילציה",
        "link": "https://vgames.co.il/",
        "favorite_show": ""
    },
    "Dean Zarhi": {
        "first_name": "Dean",
        "last_name": "Zarhi",
        "email": "dean.zarhi@e.braude.ac.il",
        "courses": "WEB, פיזיקה, מבוא לפיזיקה, מחשוב ענן, קומפילציה",
        "link": "https://vgames.co.il/",
        "favorite_show": ""
    },
    "Oshri Agronov": {
        "first_name": "Oshri",
        "last_name": "Agronov",
        "email": "oshri.agronov@e.braude.ac.il",
        "courses": "WEB, פיזיקה, מבוא לפיזיקה, מחשוב ענן, קומפילציה",
        "link": "https://vgames.co.il/",
        "favorite_show": ""
    },
    "Tal Metzger": {
        "first_name": "Tal",
        "last_name": "Metzger",
        "email": "tal.metzger@e.braude.ac.il",
        "courses": "WEB, פיזיקה, מבוא לפיזיקה, מחשוב ענן, קומפילציה",
        "link": "https://vgames.co.il/",
        "favorite_show": ""
    }
}

with open(file_path, 'w', encoding='utf-8') as file:
    json.dump(student_dict, file, indent=4)


with open(file_path, 'r', encoding='utf-8') as file:
    students_data = json.load(file)

dropdown_student = widgets.Dropdown(
    options=list(students_data.keys()),
    description='Student:',
)

text_email = widgets.Text(description='Email:', disabled=True)
text_courses = widgets.Text(description='Courses:', disabled=True)
text_link = widgets.Text(description='Link:', disabled=True)
text_fav_show = widgets.Text(description='Fav Show:', placeholder='Enter favorite show')

button_update = widgets.Button(
    description='עדכן תוכנית',
    button_style='success',
    tooltip='Update favorite show in JSON'
)
output_message = widgets.Output()

def on_student_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        selected_student = change['new']
        text_email.value = students_data[selected_student]['email']
        text_courses.value = students_data[selected_student]['courses']
        text_link.value = students_data[selected_student]['link']
        text_fav_show.value = students_data[selected_student].get('favorite_show', '')
        with output_message:
            output_message.clear_output()

dropdown_student.observe(on_student_change)

def on_update_button_clicked(b):
    selected_student = dropdown_student.value
    new_show = text_fav_show.value

    students_data[selected_student]['favorite_show'] = new_show

    with open(file_path, 'w', encoding='utf-8') as file:
        json.dump(students_data, file, ensure_ascii=False, indent=4)

    with output_message:
        output_message.clear_output()
        print(f"התוכנית של {selected_student} עודכנה בהצלחה ל-'{new_show}' ונשמרה בקובץ!")

button_update.on_click(on_update_button_clicked)

on_student_change({'type': 'change', 'name': 'value', 'new': dropdown_student.value})

display(dropdown_student)
display(text_email)
display(text_courses)
display(text_link)
display(text_fav_show)
display(button_update)
display(output_message)
```
